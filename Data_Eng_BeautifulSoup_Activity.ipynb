```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from urllib.request import urlopen
from bs4 import BeautifulSoup

# easily display plots
%matplotlib inline
```

```python
# specify the URL containing the dataset
url = "https://www.hubertiming.com/results/2023WyEasterLong" #"http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
print(url)
print(html)
```

```python
# pass the html to the BeautifulSoup function
soup = BeautifulSoup(html, 'lxml') # parse the html: take the raw html and break into python objects
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
print(soup.text)
```

```python
# a - hyperlinks
# table - tables
# tr - table rows
# th - table headers
# td - table cells
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells)) # find all the substrings where the regex matches and replace each with an empty string
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
# convert the list into a pandas df
df = pd.DataFrame(list_rows)
df.head(10)
```

Data Transformation

```python
# split the 0 column into multiple columns at the comma position using str.split
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
# remove unnecessary square brackets using strip() on column 0
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
# convert the table headers to a new pandas dataframe
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
# split column 0 into multiple columns at the comma position for all rows
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
# concatenate the two dataframes into one using the concat() method
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
# re configure the data frame so that the first row is the table header
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

the table has 583 rows and 9 columns

```python
# validation: drop all rows with missing values
df6 = df5.dropna(axis=0, how='any')
df6.info()
df6.shape
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
# clean up headers by renaming the '[Place'
df7.rename(columns={'[Place': 'Place'},inplace=True)

df7.rename(columns={' Team]': 'Team'},inplace=True)
df7.head()
```

```python
# removing the closing bracket for cells in the Team column
if 'Team' in df7.columns:
  df7['Team'] = df7['Team'].str.strip(']')
  df7.head()
```

Data Analysis and Visualization

The first question to answer: what was the average finish time (in minutes) for the runners?

```python
time_list = df7[' Time'].tolist()

# use a for loop to convert 'Time' to minutes
time_mins = []
for i in time_list:
    time_format = re.compile(':..:')
    match = re.search(time_format, i)
    if match:
        h, m, s = i.split(':')
        math = (int(h) * 3600 + int(m) * 60 + int(s))/60
    else:
        m, s = i.split(':')
        math = (int(m) * 60 + int(s))/60
    time_mins.append(math)

print(time_mins)

df7['Runner_mins'] = time_mins
df7.describe(include=[np.number])
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 15, 5


df7.boxplot(column='Runner_mins')
plt.grid(True, axis='y')
plt.ylabel('Chip Time')
plt.xticks([1], ['Runners'])
```

```python
x = df7['Runner_mins']
ax = sns.distplot(x, hist=True, kde=True, rug=False, color='m', bins=25, hist_kws={'edgecolor':'black'})
plt.show()
```

Question 3: Were there any performance differences between males and females of various age groups?

```python
f_fuko = df7.loc[df7[' Gender']==' F']['Runner_mins']
m_fuko = df7.loc[df7[' Gender']==' M']['Runner_mins']
sns.distplot(f_fuko, hist=True, kde=True, rug=False, hist_kws={'edgecolor':'black'}, label='Female')
sns.distplot(m_fuko, hist=False, kde=True, rug=False, hist_kws={'edgecolor':'black'}, label='Male')
plt.legend()
```

```python
g_stats = df7.groupby(" Gender", as_index=True).describe()
print(g_stats)
```
